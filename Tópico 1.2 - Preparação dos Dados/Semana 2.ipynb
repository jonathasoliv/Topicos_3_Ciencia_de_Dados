```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

#DataFrame original 
baseDados = pd.read_csv('game_dataset.csv', delimiter=',')
baseDados.head(10)
```

Preenchimento de Elementos Faltantes: #não podemos analisar uma base de dados faltando elementos. Precisamos preencher usando media ou mediana. A mediana ordena o vetor e seleciona o elemento do meio daquele vetor. 

```python

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='median') 
baseDados["Idade Recomendada"] = imputer.fit_transform(baseDados["Idade Recomendada"].values.reshape(-1, 1)) #reschape transforma para o vetor

imputer = SimpleImputer(missing_values=np.nan, strategy='median')
baseDados["Classificação de Usuários"] = imputer.fit_transform(baseDados["Classificação de Usuários"].values.reshape(-1, 1)) #reschape transforma para o vetor

baseDados[["Idade Recomendada","Classificação de Usuários"]]




```

Transformação de Variáveis Categóricas: #separa as colunas nos tipos de genero, plataforma e lançamento com os tipos preenchendo as colunas em 0 e 1. A linha 0 é quando não tem informação e a linha 1 tem informação

```python

baseDados = pd.get_dummies(baseDados, columns=['Gênero', 'Plataforma', 'Lançamento no Brasil'], prefix=['Gênero', 'Plataforma', 'Lançamento no Brasil'])
baseDados.head(8)
```

Normalização de Dados: #estabelecer um intervalo entre 0 e 1 para as colunas Preco nao ter tanto peso sobre a coluna Vendas Globais no momento da analise

```python
scaler = MinMaxScaler() #transforma as caracteristicas (colunas) na base para ter um intervalo especifico de valores entre 0 e 1  
baseDados[['Vendas Globais','Preço']] = scaler.fit_transform(baseDados[['Vendas Globais','Preço']])

baseDados[["Vendas Globais","Preço"]]


#baseDados.head()
```
